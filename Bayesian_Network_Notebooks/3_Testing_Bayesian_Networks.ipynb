# Evaluation Metrics:
There are 10 different metrics used to evaluate a BN network, ranging from Machine Learning metrics to more specialised metrics for BNs:
- **Precision** -  rate of correct edges over all edges
discovered.
- **Recall** - rate of edges discovered over all edges in the
true graph.

- **F1** - harmonic mean of the precision and recall
- **Structural Hamming Distance** (SHD) - minimum number of edge insertions,
deletions, and arc reversals needed to transform the learned graph into the true graph (**Lower** value is better.)
- **DAG Dissimilarity Metric** (DDM) - a score that ranges from −∞ to 1, where the score of 1
indicates perfect agreement between the two graphs. The score moves to −∞ the stronger the
dissimilarity is between the two graphs. (**Higher** value is better)
- **Balanced Scoring Function** (BSF) - ranges from -1 to 1, where a score of -1 corresponds to the worst possible
graph (i.e., the reverse of the true graph), a score of 1 corresponds to a perfect match between
the learned and the true graphs, and a score of 0 corresponds to a learned graph that is
equivalent to the score generated by an empty or a fully connected graph. (**Higher** value is better)
- **Log Likelihood** (LL) - how well the model fits the input data, how likely is the model to have generated the input data we have observed. (A **higher** log-likelihood value means the model predicts the observed data well)
-  **Bayesian Information Criterion** (BIC) - takes the model fitting score and adds dimensionality penalty - takes into consideration sample size, number of free parameters, how well the model fits the data, accuracy of parameters that you create with each additional parent. (**Lower** value is better)

- **Free Parameters** - dimensionality or complexity of the model.



# Algorithms:

- **HC** – hill-climbing search, starts from empty graph performing a hill-climbing by exploring arc additions, reversals and deletions, moving to the DAG that maximises the BIC score.
- **SaiyanH** – hybrid structure learning algorithim. Starts with a dependency function producing an undirect graph that can be viewed as an extended version of the maximum spanning graph, ensuring that each variable associates with at least one edge. It then obtains a DAG using a combination of constraint-based, score-based and interventional impact rules. The DAG is then provided as input to Tabu search, with the restriction not to delete edges that lead to disjoint subgraphs or disjoint nodes.
- **TABU** – classic score-based TABU search. When it reaches a local maximum (highest point in a limited space), it attempts to decrease the BIC score and uses HC on each iteration of a set of variables to escape this local maximum.
- **MAHC** (Model-Averaging Hill-Climbing)  - The algorithm starts by pruning the search space of graphs, where the pruning strategy can be viewed as an aggressive version of the pruning strategies that are typically applied to combinatorial optimisation structure learning problems. It then performs model averaging in the hillclimbing search process and moves to the neighbouring graph that maximises the objective function, on average, for that neighbour and over all its valid neighbours.
- **GES** (Greedy Equivalence Search) - a greedy algorithm that explores the space of Markov equivalence classes [8]. It starts with an ‘insert’ phase that explores edge additions that increase the objective score the most, and then moves to the ‘delete’ phase in an analogous fashion until a final maximum scoring graph is produced.

- **My knowledge** (Not an algorithm) - I created a Bayesian Network based upon research and knowledge determining what variables have a causal effect on others for the Obesity dataset.
